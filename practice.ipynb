# Разработайте алгоритм на Python для коррекции цвета изображения, переводя его в пространство CIE XYZ и обратно

```python
from skimage import io
from skimage import color
from skimage import data
from pylab import *

img = io.imread("kafka.jpg")

img_xyz = color.rgb2xyz(img)

img_rgb = color.xyz2rgb(img)

figure(0)
io.imshow(img)

figure(1)
io.imshow(img_xyz)

figure(2)
io.imshow(img_rgb)
```
